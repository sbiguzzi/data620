## Assignment Steps
We first import the libraries necessary for this assignment: `networkx` as "net" and `matplotlib.pyplot` as "plt". We also tell python to show the plots inline with the command `%matplotlin inline`

```python
import networkx as net
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

The next step is to instantiate the Graph class from the `networkx` library and assingning it to the variable _G_.

```python
G = net.Graph()
```

We assign the nodes and labels for those nodes in the dictionary _labels_. __Nodes__ are the points on your notwork graph plot. In the example below node _0_ is associated with label _'Stefano'_, node _1_ with _Ian_ and so forth. In our example, we were asked to recreate a 9 node graph plot (see Figure 1).

```python
labels = {
    0:'Stefano',
    1:'Ian',
    2:'Dennis',
    3:'Karim',
    4:'Ed',
    5:'Fernando',
    6:'Garth',
    7:'Heather',
    8:'Ike',
    9:'Jane',
}
```

We have to assign the proper connections for each node so that the final graph is the same as Figure 1. __Edges__ are the different connections each node can have. In our example below edge _(0,1)_ means that there is a connection between _Stefano_ (node 0) and _Ian_ (node 1), edge _(0,2)_ mean there's a connection between _Stefano_ and _Dennis_, and edge _(0,3)_ means there is a connection between _Stefano_ and _Karim_. We use the `add_edges_from()` method to assign the predetermined edges to our graph _G_.

```python
edges = [(0,1),(0,2),(0,3),(0,5),(1,3),(1,4),(1,6),(2,3),(2,5),(3,4),(3,5),(3,6),(4,6),(5,6),(5,7),
         (6,7),(7,8),(8,9)]
G.add_edges_from(edges)
```

We reassign graph _G_ to a new graph _G1_ using the method `relabel_node()`, by passing through the original graph `G` and the node labels `labels` as our parameters.

```python
G1 = net.relabel_nodes(G,labels)
```

Finally, we use the `net.draw()` method, with parameters `G1` and `with_labels=1`, to instanciate the graph and the `plt.show()` method to show the graph (One thing we haven't figured out is how to make the graph stay fixed every time the draw method is run).

```python
net.draw(G1,with_labels=1)
plt.show()
```
